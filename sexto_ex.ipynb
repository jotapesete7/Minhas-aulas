```python
import numpy as np
import matplotlib.pyplot as plt
# Criando as listas com altura e peso de 20 pessoas
peso = [60.2, 73.5, 64.0, 68.1, 87.0, 75.4, 70.0, 76.8, 81.2, 58.9, 69.7]
altura = [1.65, 1.75, 1.68, 1.72, 1.90, 1.80, 1.77, 1.82, 1.87, 1.64, 1.74]
#Calculando a corelaçao entre altura e peso
corr_coef = np.corrcoef(peso, altura)[0,1]
#Plotando o gráfico de dispersão com duas variáveis
plt.scatter(altura, peso)
plt.xlabel('Altura (m)')
plt.ylabel('Peso (kg)')
plt.title(f'correlação:{corr_coef:.2f}')
plt.show()
```
