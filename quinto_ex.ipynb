```python
import matplotlib.pyplot as plt
#Dados
regioes = ['norte', 'nordeste', 'centro-oeste', 'sudeste', 'sul']
vendas = [1500,2800,1800,3000,2700]
#Criar gtáfico de barras
plt.bar(regioes, vendas)
#Configurar títulos e rótulos dos eixos
plt.title('vendas novo produto por regiões')
plt.xlabel('regioes')
plt.ylabel('quantidade de vendas')
#Mostrar gráfico
plt.show()




```
